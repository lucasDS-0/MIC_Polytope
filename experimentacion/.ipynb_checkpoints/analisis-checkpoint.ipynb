```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_resultados = pd.read_csv("resultados/resultados-mem-size.csv");
df_resultados
```

```python
mem_size = df_resultados.copy();

fig = mem_size.boxplot(by='mem-size', column=['error'], grid=True)
#fig = sns.scatterplot(data=mem_size, x='mem-size', y='error');
#plt.legend(title='Memory size', loc='upper left')
plt.show(fig)
plt.savefig("graficos/mem-size.svg")
```
